### Preparing Data for Linear Regression Modeling

```python
from sklearn.linear_model import LinearRegression
```

```python
reg= LinearRegression()
```

```python
labels=data['price']
conv_dates=[1 if values==2014 else 0 for values in data.date]
data['date']= conv_dates
train1=data.drop(['id', 'price'], axis=1)
```

### Splitting Data and Training Linear Regression Model

- The code splits the data into training and testing sets using train_test_split from scikit-learn, with 90% of the data used for training and 10% for testing. It then fits a linear regression model (reg) to the training data.

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train,  y_test = train_test_split(train1, labels, test_size=0.10, random_state=2)
```

```python
reg.fit(x_train, y_train)
```

### Training Gradient Boosting Regressor

- The code initializes a Gradient Boosting Regressor (clf) from the scikit-learn ensemble module, with specified hyperparameters such as the number of estimators, maximum depth, minimum samples split, learning rate, and loss function. It then fits the regressor to the training data.

```python
from sklearn import ensemble
clf = ensemble.GradientBoostingRegressor(n_estimators=400, max_depth=5, 
                                         min_samples_split=2, 
                                        learning_rate=0.1,
                                        loss='squared_error')
```

```python

clf.fit(x_train,y_train)
```

```python
clf.score(x_test, y_test)
```

### Assessing Model Performance and Tuning Parameters
- The code evaluates the performance of the model on the test set using the score method. It then calculates the loss on the test set for each stage of the gradient boosting process using staged_predict and stores the results in an array t_sc. Finally, it generates an array testsc representing the number of estimators.

```python
params={'n_estimators': 400}
t_sc= np.zeros((params['n_estimators']), dtype =np.float64)
```

```python
y_pred =reg.predict(x_test)
```

```python
for i , y_pred in enumerate(clf.staged_predict(x_test)):
    t_sc[i]= clf.loss_(y_test, y_pred)
```

```python
testsc=np.arange((params['n_estimators']))+1
```

13. **Visualizing Model Deviance**
- The plot compares the training set deviance (blue line) and test set deviance (red line) as the number of estimators increases. This visualization helps in assessing the model's training progress and identifying overfitting or underfitting tendencies.

```python
#
plt.figure(figsize=(12,6))
plt.subplot(1,2,1)
plt.plot(testsc, clf.train_score_, 'b-', label='Set dev train')
plt.plot(testsc, t_sc, 'r-', label='set dev test')
```
